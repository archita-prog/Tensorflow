```python
import gensim
import pandas as pd
```

```python
df = pd.read_json(r"C:\Users\Archita Shrivastava\Downloads\reviews_Cell_Phones_and_Accessories_5.json\Cell_Phones_and_Accessories_5.json", lines =True)
df.head()
```

```python
df.shape
```

```python
df.reviewText[0]
```

```python
review_text = df.reviewText.apply(gensim.utils.simple_preprocess)
review_text
```

```python
model = gensim.models.Word2Vec(
    window = 10,
    min_count = 2,
    workers= 4
)
```

```python
model.build_vocab(review_text, progress_per = 1000)
```

```python
model.epochs
```

```python
model.corpus_count
```

```python
model.train(review_text, total_examples = model.corpus_count,epochs = model.epochs )
```

```python
model.save("./word2vec-amazon-cell-accessories-reviews-short.model")
```
